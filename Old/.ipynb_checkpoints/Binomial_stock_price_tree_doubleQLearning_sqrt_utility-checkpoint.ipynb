```python
# Libraries
import gym
from envs.binomial_tree import BinomialTree, decode_action 
from envs import plotting
import numpy as np
import pandas as pd
import math
from collections import defaultdict
import itertools
import matplotlib.pyplot as plt
```

```python
actions = np.arange(0, 1.01, step=0.1)   #vector of actions
#actions = np.array([0, 0.68, 1])
print("Actions (Investment in risky asset):", actions)
lower = 90
upper = 110
delta_bin = 20
wealth_bins = [0] + np.arange(lower, upper+1, delta_bin).tolist() + [float('Inf')]  # +1 as upper limit is not included
```

**Double Q-Learning Algorithm**\
Source: https://rubikscode.net/2020/01/20/double-q-learning-python/

```python
def double_q_learning(env, num_episodes, alpha=0.05, eps=0.1, gamma=1):

    q_a_table = defaultdict(lambda: np.zeros(env.action_space.n)) 
    q_b_table = defaultdict(lambda: np.zeros(env.action_space.n)) 
    rewards = np.zeros(num_episodes)
    
    # Keeps track of useful statistics 
    stats = plotting.EpisodeStats( 
        episode_lengths = np.zeros(num_episodes), 
        episode_rewards = np.zeros(num_episodes)) 

    
    for episode in range(0, num_episodes):       
        # Reset the environment
        state = env.reset()
        
        # Initialize variables
        done = False
        t = 0

        while not done:
            t = t+1
            
            # Pick action a....
            if np.random.rand() < eps:
                action = env.action_space.sample()
            else:
                q_table = q_a_table[state] + q_b_table[state]
                max_q = np.where(np.max(q_table) == q_table)[0]
                action = np.random.choice(max_q)

            # ...and get r and s'    
            next_state, reward, done, _ = env.step(action)
            
            # Update statistics 
            stats.episode_rewards[episode] += reward 
            stats.episode_lengths[episode] = t
            
            # Update(A) or Update (B)
            if np.random.rand() < 0.5:
                # If Update(A)
                q_a_table[state][action] += alpha * (reward + gamma * q_b_table[next_state][np.argmax(q_a_table[next_state])] - q_a_table[state][action])

            else:
                # If Update(B)
                q_b_table[state][action] = alpha * (reward + gamma * q_a_table[next_state][np.argmax(q_b_table[next_state])] - q_b_table[state][action])

            state = next_state
            rewards[episode] += reward  
            
        if episode % 10000 == 0:
            print("Episode: {}".format(episode))
            
    return rewards, q_a_table, q_b_table, stats
```

```python
# Square root utility function
#up_prob, up_ret, down_ret, r, T, dt, V_0, actions, wealth_bins, utility
env = BinomialTree(up_prob=4/9, up_ret=1, down_ret=-1/2, r=0, T=2, dt=1, V_0=100, actions=actions, wealth_bins=wealth_bins, utility="sqrt")
```

```python
dq_reward, q_a_table, q_b_table, stats = double_q_learning(env, num_episodes=150000)
```

```python
plotting.plot_episode_stats(stats) 
```

```python
# State-actions values for each state
for key in q_a_table:
    print("Key: {}, Q(s, a): {}".format(key, q_a_table[key]))
    print("Best Action (Investment in risky asset):", decode_action(np.argmax(q_a_table[key]), actions))
```
